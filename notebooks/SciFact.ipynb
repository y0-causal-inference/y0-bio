# SciFact

Due to the rapid growth in the scientific literature, there is a need for automated systems to assist researchers and the public in assessing the veracity of scientific claims. To facilitate the development of systems for this task, we introduce SciFact, a dataset of 1.4K expert-written claims, paired with evidence-containing abstracts annotated with veracity labels and rationales.



## Example SciFact Question

| Claim	| Evidence	| Decision |
|------|------------|----------|
|Prescribed exercise training improves quality of life.	| At 3 months, usual care plus exercise training led to greater improvement in the KCCQ overall summary score (mean, 5.21; 95% confidence interval, 4.42 to 6.00) compared with usual care alone (3.28; 95% confidence interval, 2.48 to 4.09). |	SUPPORTS |
| Patients with microcytosis and higher erythrocyte count are more vulnerable to severe malarial anaemia. |	The increased erythrocyte count and microcytosis in children homozygous for alpha(+)-thalassaemia may contribute substantially to their protection against SMA. |	REFUTES |

## Scoring
We evaluate models on their ability to identify relevant evidence supporting and / or refuting a given claim. We evaluate at both the *abstract level* (i.e. how many research abstracts did the model correctly identify), and the *sentence level* (i.e. how well did the model identify the relevant evidence from the cited abstracts). For more information, see Section 4 of the [paper](https://www.aclweb.org/anthology/2020.emnlp-main.609/). For more information, see this [detailed example](https://github.com/allenai/scifact/blob/master/doc/evaluation.md), which shows how evaluation would be performed on a formatted example submission.

For more information on any of the leaderboard columns, hover over the desired column or click the *expand* button on the upper right of the leaderboard to display full column names.

You can download the [source of the evaluator](https://github.com/allenai/scifact-evaluator) to see how it works for yourself.

## Predictions Format

Your predictions should be output as a jsonl file, with one row per claim. A prediction for a single claim should be formatted as follows. Two minor notes:

The value of the `"id"` field (e.g. 52) should be stored as an `int`.
The keys in the evidence object (e.g. `"11"`, `"16"`) should be stored as strings, because JSON requires key names to be strings.

```
{
  "id": 52,
  "evidence": {
      "11": {
          "sentences": [ 1, 11, 13 ],     // Predicted rationale sentences.
          "label": "SUPPORT"              // Predicted label.
      },
      "16": {
        "sentences": [18, 20 ],
        "label": "REFUTES"
      }
  }
}
```

See the [detailed instructions](https://github.com/allenai/scifact/blob/master/doc/evaluation.md) in the SciFact repo for more examples.

## Rules

* For abstract-level predictions, *only the first 3* predicted rationale sentences will be counted; the rest will be discarded.
* For sentence-level predictions, all predicted rationale sentences will be counted.
* For more details, see [here](https://github.com/allenai/scifact/blob/master/doc/evaluation.md).

# Convert claims to Biological Expression Language using Reach

## Read training set of claims

````python
from indra.sources import reach
#rp = reach.process_text('Patients with microcytosis and higher erythrocyte count are more vulnerable to severe malarial anaemia.')
````

````python
import json, pathlib
claims_dir = pathlib.Path('/Users/zuck016/Projects/CausalInference/MAA/scifact/data/data')
claims_train_file = claims_dir / 'claims_train.jsonl'
def read_claims(claims_file):
    claims = []
    for line in claims_file.read_text().split('\n'):
        if len(line) > 0:
            claims.append(json.loads(line))
    return claims

claims_train = read_claims( claims_train_file)
````

## Send claim to local reach server

````python
def parse_claims( claims ):
    for claim in claims:
        rp = reach.process_text(claim['claim'],  url=reach.local_text_url)
        claim['statements_local'] = rp.statements
    return claims

claims_train = parse_claims( claims_train )
````

````python
from indra.assemblers.pybel.assembler import PybelAssembler

statements_local = []
indra_claims_local =  []
for claim in claims_train:
    if len(claim['statements_local']) >0:
        pba = PybelAssembler(claim['statements_local'])
        claim['bel_local'] = pba.make_model()
        statements_local += claim['statements_local']
        print(claim)
        indra_claims_local.append(claim)
pba = PybelAssembler(statements_local)
belgraph_local = pba.make_model()
belgraph_local
````

# Find the PMID for each paper in the corpus by title

## Read in the corpus

````python
corpus_file = claims_dir / 'corpus.jsonl'
corpus = []
for line in corpus_file.read_text().split('\n'):
    if len(line) > 0:
        corpus.append(json.loads(line))
corpus
````

## Look up the title in Entrez

````python
from Bio import Entrez
````

````python
Entrez.email = "jeremy.zucker@pnnl.gov" 
records = {}# Always tell NCBI who you are
for article in corpus:
    title = article['title']
    handle = Entrez.esearch(db="pubmed", term=f"{title}[title]", retmax="1" )
    records[title] = Entrez.read(handle)
    print(f'{title}: {records[title]["IdList"]}')
````

## Assign the pmid back to the article

````python
for article in corpus:
    title = article['title']
    article['pmid'] = records[title]['IdList']

````

````python
import json
with open(claims_dir / 'corpus_pmid.json', 'w') as fp:
    json.dump(corpus, fp, indent=2)
````
