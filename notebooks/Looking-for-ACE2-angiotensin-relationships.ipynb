```python
def get_statements_by_subject_and_object( subject, obj, emmaa_covid_graph):
    statements = []
    for u, v, data in emmaa_covid_graph.edges(subject, data=True):
        if v in obj:
            #print(emmaa_covid_graph.edge_to_bel(u, v, data))
            if 'evidence' in data:
                statements.append([u,v,data])
    return statements
sars2_endocytosis = get_statements_by_subject_and_object( sars2, endocytosis, emmaa_covid_graph )
sars2_endocytosis
```

## Find all INDRA statements associated with a publication

```python
def get_statements_by_pubmed( pubmed, emmaa_covid_graph=emmaa_covid_graph ):
    statements = []
    for u, v, data in emmaa_covid_graph.edges(data=True):
        if  ('citation' in data) and (
             data['citation']['namespace'] == 'pubmed') and (
             data['citation']['identifier'] ==  pubmed):
                #print(emmaa_covid_graph.edge_to_bel(u, v, data))
                if 'evidence' in data:
                    #print(data['citation']['namespace'], data['citation']['identifier'])
                    #print(data['evidence'])
                    statements.append([u,v,data])
    return statements
        
```

```python
import networkx as nx
G = nx.DiGraph()
G.add_edge("x", "a", capacity=3.0)
G.add_edge("x", "b", capacity=1.0)
G.add_edge("a", "c", capacity=3.0)
G.add_edge("b", "c", capacity=5.0)
G.add_edge("b", "d", capacity=4.0)
G.add_edge("d", "e", capacity=2.0)
G.add_edge("c", "y", capacity=2.0)
G.add_edge("e", "y", capacity=3.0)
```

```python
df = pd.DataFrame([
    (str(u), str(v))
    for u, v, data in emmaa_covid_graph.edges(data=True)
    if isinstance(u, pybel.dsl.Protein) and isinstance(v, pybel.dsl.Protein)
])
df.drop_duplicates().head()
```

```python
cut_value, partition = nx.minimum_cut(G, "x", "y")
reachable, non_reachable = partition
```

```python
cutset = set()
for u, nbrs in ((n, G[n]) for n in reachable):
    cutset.update((u, v) for v in nbrs if v in non_reachable)
print(sorted(cutset))

cut_value == sum(G.edges[u, v]["capacity"] for (u, v) in cutset)

```

```python
from networkx.algorithms.flow import shortest_augmenting_path
cut_value == nx.minimum_cut(G, "x", "y", flow_func=shortest_augmenting_path)[0]
```
